```python
import funciones as fn


###################################
# Funciones principales del juego #
###################################

# Menu de dificultad
def dificultad():
    fn.submenu()
    dificultad = fn.valida(1, 4)
    if dificultad == 1:
        fn.musica_fondo_facil()
        return 20
    elif dificultad == 2:
        fn.musica_fondo_medio()
        return 12
    elif dificultad == 3:
        fn.musica_fondo_dificil()
        return 5
    else:
        print("\nVolviendo", end="", flush=True)
        for _ in range(3):
            fn.time.sleep(0.4)
            print(".", end="", flush=True)
        fn.time.sleep(0.5)
        print("\n")  
        return None # Volver al menÃº principal si la opciÃ³n no es vÃ¡lida

# Guardar estadÃ­sticas
def guardar_stats(estadisticas_jugador):
    bbdd_guessthenumber_act = fn.pd.DataFrame(estadisticas_jugador, columns=["Modo", "Nombre", "NÃºmero a adivinar", "Intentos", "Resultado", "Fecha y hora"])
    # Comprobar si el archivo ya existe para aÃ±adir los nuevos datos
    if fn.os.path.exists("estadisticas_jugador.xlsx"):
        bbdd_guessthenumber_ant = fn.pd.read_excel("estadisticas_jugador.xlsx")
        bbdd_guessthenumber = fn.pd.concat([bbdd_guessthenumber_ant, bbdd_guessthenumber_act], ignore_index=True)
    # Si no existe, crear uno nuevo
    else:
        bbdd_guessthenumber = bbdd_guessthenumber_act
    bbdd_guessthenumber.to_excel("estadisticas_jugador.xlsx", index=False)
    return

# Modo solitario
def modo_solitario():
    # Establecer nÃºmero de intentos segÃºn dificultad
    intentos = dificultad()
    if intentos is None: # Volver al menÃº principal si no hay intentos guardados (opciÃ³n 4)
        return
    # Datos que se van a guardar
    modo = "Solitario"
    numero_a_adivinar = fn.rdm.randint(1, 1000)
    nombre_jugador = input("Introduce tu nombre para guardar tu progreso: ")
    estadisticas_jugador = []
    
    # Frases aleatorias para pistas
    pistas_mayor = [
        "\nğŸ”º Â¡MÃ¡s arriba, mÃ¡s arriba!\n",
        "\nğŸ“ˆ Sube un poco mÃ¡s, Â¡casi llegas!\n",
        "\nğŸ˜ El nÃºmero es mÃ¡s grande...\n",
        "\nğŸš€ Necesitas apuntar mÃ¡s alto.\n",
        "\nğŸ§—â€â™‚ï¸ Piensa en algo mÃ¡s grande.\n"
    ]

    pistas_menor = [
        "\nğŸ”» Â¡Demasiado alto, bÃ¡jale un poco!\n",
        "\nğŸ“‰ Ups, te pasaste. Prueba un nÃºmero menor.\n",
        "\nğŸ˜… No tan alto, intenta mÃ¡s bajo.\n",
        "\nğŸ‚ Baja un poco, que te pasaste.\n",
        "\nğŸœ El nÃºmero es mÃ¡s pequeÃ±o que este.\n"
    ]

    # Bucle de intentos para adivinar el nÃºmero
    for i in range(intentos):
        numero_introducido = fn.valida_numero(nombre_jugador)
        if numero_introducido < numero_a_adivinar:
            print(fn.rdm.choice(pistas_mayor))
        elif numero_introducido > numero_a_adivinar:
            print(fn.rdm.choice(pistas_menor))
        else:
            print(f"\nğŸ‰ Â¡Has adivinado el nÃºmero en {i+1} intentos!\n")
            print(f"\nğŸ˜„Â¡SÃ­ seÃ±or! Era {numero_a_adivinar}. Â¡QuÃ© punterÃ­a tienes {nombre_jugador}!\n")
            fn.sonido_victoria()
            resultado = "Victoria"
            fecha_hora_actual = fn.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            estadisticas_jugador.append((modo, nombre_jugador, numero_a_adivinar, i+1, resultado, fecha_hora_actual))
            guardar_stats(estadisticas_jugador)
            return
    else:
        print(f"\nğŸ˜¢ Se acabaron los intentos. El nÃºmero era {numero_a_adivinar}. Pero oye, la intenciÃ³n es lo que cuenta... ğŸ˜\n")
        print(f"\nğŸ’ª Â¡No te rindas {nombre_jugador}! La prÃ³xima vez seguro lo consigues.\n")
        fn.sonido_derrota()
        resultado = "Derrota"
        fecha_hora_actual = fn.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        estadisticas_jugador.append((modo, nombre_jugador, numero_a_adivinar, i+1, resultado, fecha_hora_actual))
        guardar_stats(estadisticas_jugador)
        return
    
# Modo multijugador
def modo_multijugador():
    # Establecer nÃºmero de intentos segÃºn dificultad
    intentos = dificultad()
    if intentos is None: # Volver al menÃº principal si no hay intentos guardados (opciÃ³n 4)
        return
    # Datos que se van a guardar
    modo = "Multijugador"
    nombre_jugador1 = input("Jugador 1, introduce tu nombre: ")
    nombre_jugador2 = input("Jugador 2, introduce tu nombre: ")
    numero_a_adivinar_jugador1 = fn.valida_numero_oculto(nombre_jugador1)
    estadisticas_jugador = [] 

    # Frases aleatorias para pistas
    pistas_mayor = [
        "\nğŸ”º Â¡MÃ¡s arriba, mÃ¡s arriba!\n",
        "\nğŸ“ˆ Sube un poco mÃ¡s, Â¡casi llegas!\n",
        "\nğŸ˜ El nÃºmero es mÃ¡s grande...\n",
        "\nğŸš€ Necesitas apuntar mÃ¡s alto.\n",
        "\nğŸ§—â€â™‚ï¸ Piensa en algo mÃ¡s grande.\n"
    ]

    pistas_menor = [
        "\nğŸ”» Â¡Demasiado alto, bÃ¡jale un poco!\n",
        "\nğŸ“‰ Ups, te pasaste. Prueba un nÃºmero menor.\n",
        "\nğŸ˜… No tan alto, intenta mÃ¡s bajo.\n",
        "\nğŸ‚ Baja un poco, que te pasaste.\n",
        "\nğŸœ El nÃºmero es mÃ¡s pequeÃ±o que ese.\n"
    ]

    # Bucle de intentos para adivinar el nÃºmero
    for i in range(intentos):
        numero_introducido_jugador2 = fn.valida_numero(nombre_jugador2)
        if numero_introducido_jugador2 < numero_a_adivinar_jugador1:
            print(fn.rdm.choice(pistas_mayor))
        elif numero_introducido_jugador2 > numero_a_adivinar_jugador1:
            print(fn.rdm.choice(pistas_menor))
        else:
            print(f"\nğŸ‰ Â¡Has adivinado el nÃºmero en {i+1} intentos!\n")
            print(f"\nğŸ˜„ Â¡SÃ­ seÃ±or! Era {numero_a_adivinar_jugador1}. Â¡{nombre_jugador1} no ha podido contigo, {nombre_jugador2}!\n")
            fn.sonido_victoria()
            resultado = "Victoria"
            fecha_hora_actual = fn.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            estadisticas_jugador.append((modo, nombre_jugador2, numero_a_adivinar_jugador1, i+1, resultado, fecha_hora_actual))
            guardar_stats(estadisticas_jugador)
            return
    else:
        print(f"\nğŸ˜¢ Se acabaron los intentos. El nÃºmero era {numero_a_adivinar_jugador1}. Pero oye, la intenciÃ³n es lo que cuenta... ğŸ˜\n")
        print(f"\nğŸ’ª Â¡Vaya nÃºmero te ha puesto {nombre_jugador1}! La prÃ³xima vez seguro lo consigues {nombre_jugador2}.\n")
        fn.sonido_derrota()
        resultado = "Derrota"
        fecha_hora_actual = fn.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        estadisticas_jugador.append((modo, nombre_jugador2, numero_a_adivinar_jugador1, i+1, resultado, fecha_hora_actual))
        guardar_stats(estadisticas_jugador)
        return  

# EstadÃ­sticas
def estadistica():
    print("\nCargando", end="", flush=True)
    for _ in range(3):
        fn.time.sleep(0.4)
        print(".", end="", flush=True)
    fn.time.sleep(0.5)
    print("\n")
    # Comprobar si el archivo de estadÃ­sticas existe
    if fn.os.path.exists("estadisticas_jugador.xlsx"):
        bbdd_guessthenumber = fn.pd.read_excel("estadisticas_jugador.xlsx")
        texto = "ğŸ“Š ESTADÃSTICAS DE JUEGO ğŸ“Š"
        print(texto.center(90))
        print("=" * 90)
        print(bbdd_guessthenumber.to_string(index=False))
        print("=" * 90 + "\n")
        return
    # Si no existe, informar al jugador
    else:
        print("No hay estadÃ­sticas guardadas.")
        return
```
