## Ejercicio 10
Crea un juego de adivinanza para dos jugadores:
- El primer jugador introduce un n√∫mero comprendido entre 1 y 1000.
- El segundo jugador debe adivinarlo en un m√°ximo de 10 intentos.

Tras cada intento, el programa indicar√° si el n√∫mero buscado es menor o mayor que el introducido. Si acierta, mostrar√° un mensaje de √©xito; si no lo logra en los 10 intentos, mostrar√° un mensaje de que ha perdido.

```python
jugador1 = int(input("Introduce un n√∫mero del 1 al 1000: "))

if 1 <= jugador1 <= 1000:
    for intento in range(11):
        if intento < 10:
            jugador2 = int(input("Adivina el n√∫mero: "))
            if jugador2 > jugador1:
                print("El n√∫mero es menor.")
            if jugador2 < jugador1:
                print("El n√∫mero es mayor.")
            if jugador2 == jugador1:
                print("Has ganado.")
                break
        else: 
            intento == 10
            print("Has perdido.")
else:
    print("Introduce un n√∫mero v√°lido.")
```

## Ejercicio 3
Crea un gr√°fico que muestre si hay m√°s aprobados (notas ‚â• 5) o suspensos (notas 
< 5).


```python
import matplotlib.pyplot as plt
import seaborn as sns

bbdd = pd.read_excel("notas.xlsx")
bbdd["Aprobado"] = bbdd["Notas"].apply(lambda x: "Aprobado" if x >= 5 else "Suspenso")

sns.countplot(data=bbdd, x="Aprobado", palette="pastel")
plt.title("Notas")
plt.ylabel("N√∫mero de alumnos")
plt.xlabel("Resultado")
# S√≥lo n√∫meros enteros
import matplotlib.ticker as ticker
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
```
