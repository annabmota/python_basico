```python
# Validar opci√≥n y/o dificultad
def valida(minimo, maximo):
    mensaje = f"Elige una opci√≥n entre {minimo} y {maximo}: "
    while True:
        try:
            opcion = int(input(mensaje))
            if minimo <= opcion <= maximo:
                return opcion
            else:
                mensaje = f"Opci√≥n no v√°lida. Debe estar entre {minimo} y {maximo}: "
        except ValueError:
            mensaje = f"Valor no v√°lido. Introduce un n√∫mero entre {minimo} y {maximo}: "

# Menu de dificultad
def submenu():
    print("1. F√°cil (20 intentos) \n2. Medio (12 intentos) \n3. Dif√≠cil (5 intentos)")
    dificultad = valida(1, 3)
    if dificultad == 1:
        intentos = 20
    elif dificultad == 2:
        intentos = 12
    else:
        intentos = 5
    return intentos

# Menu de opciones
def menu():
    print("=== Adivina el n√∫mero === \n1. Modo solitario \n2. Modo multijugador \n3. Estad√≠stica \n4. Salir")
    opcion = valida(1, 4)
    if opcion == 1:
        modo_solitario()
    elif opcion == 2:
        modo_multijugador()
    # elif opcion == 3:
        # estadistica()
    else:
        print("¬°Hasta luego!")
    return
```

# **Modo solitario**
Modo  solitario  supone  que  el  n√∫mero  a  adivinar  es  generado  aleatoriamente  por  el ordenador. El n√∫mero a adivinar debe ser entre 1 y 1000. 

```python
guess.modo_solitario()
```

```python
# Modo solitario
import random as rdm
def modo_solitario():
    modo = "Solitario"
    intentos = submenu()
    numero_a_adivinar = rdm.randint(1, 1000) # Validar que el n√∫mero est√© entre 1 y 1000
    nombre_jugador = input("Introduce tu nombre para guardar tu progreso: ")
    estadisticas_jugador = []
    for i in range(intentos):
        numero_introducido = int(input(f"{nombre_jugador}, adivina el n√∫mero entre 1 y 1000: "))
        if numero_introducido < numero_a_adivinar:
            print("El n√∫mero es mayor.")
        elif numero_introducido > numero_a_adivinar:
            print("El n√∫mero es menor.")
        else:
            print(f"üéâ ¬°Has adivinado el n√∫mero en {i+1} intentos!")
            estadisticas_jugador.append((modo, nombre_jugador, numero_a_adivinar, i+1))
            guardar_stats(estadisticas_jugador)
            return
    else:
        print(f"üò¢ Se acabaron los intentos. El n√∫mero era {numero_a_adivinar}.")
        estadisticas_jugador.append((modo, nombre_jugador, numero_a_adivinar, i+1))
        guardar_stats(estadisticas_jugador)
        return
```

# **Modo multijugador**

intenta adivinarlo.

```python
# Modo multijugador
def modo_multijugador():
    modo = "Multijugador"
    intentos = submenu()
    nombre_jugador1 = input("Jugador 1, introduce tu nombre: ")
    nombre_jugador2 = input("Jugador 2, introduce tu nombre: ")
    numero_a_adivinar_jugador1 = int(input((f"{nombre_jugador1}, introduce el n√∫mero a adivinar (entre 1 y 1000): ")))
    estadisticas_jugador = [] # Hacer que los n√∫meros no se vean al escribir
    for i in range(intentos):
        numero_introducido_jugador2 = int(input(f"{nombre_jugador2}, adivina el n√∫mero entre 1 y 1000: "))
        if numero_introducido_jugador2 < numero_a_adivinar_jugador1:
            print("El n√∫mero es mayor.")
        elif numero_introducido_jugador2 > numero_a_adivinar_jugador1:
            print("El n√∫mero es menor.")
        else:
            print(f"üéâ ¬°Has adivinado el n√∫mero en {i+1} intentos!")
            estadisticas_jugador.append((modo, nombre_jugador2, numero_a_adivinar, i+1))
            guardar_stats(estadisticas_jugador)
            return
    else:
        print(f"üò¢ Se acabaron los intentos. El n√∫mero era {numero_a_adivinar_jugador1}.")
        estadisticas_jugador.append((modo, nombre_jugador2, numero_a_adivinar, i+1))
        guardar_stats()
        return
```

# **Estad√≠stica**
Se gane o se pierda se pide el nombre del jugador y se guarda esta informaci√≥n junto 
con el resultado de la partida en un fichero Excel. 
La opci√≥n estad√≠stica nos mostrara los datos almacenados en Excel. 

```python
# Guardar estad√≠sticas
def guardar_stats(estadisticas_jugador):
    import openpyxl
    import pandas as pd
    bbdd_guessthenumber = pd.DataFrame(estadisticas_jugador, columns=["Modo", "Nombre", "N√∫mero a adivinar", "Intentos"])
    bbdd_guessthenumber.to_excel("estadisticas_jugador.xlsx", index=False)
    return
```
